Question 9. What is the main difference between Simple and Multiple Linear Regression?

Answer:-> The main difference between SLR and MLR is as follows:

1. On the basis of number of known features taken into account:

In SLR, the number of known features that are taken into account to compute about the unknown feature is 1 only.

In MLR, the number of known features that are taken into account to compute about the unknown feature are generally 1.

2. On the basis of equation:

In SLR, the equation is, y = b0+b1x1

In MLR, the equation is, y = b0+b1x1+b2x2+...+bnxn

3. On the basis of graph:

In SLR, the graph obtained is straight line.

In MLR, the graph obtained is a hyperplane.

4. Example:

SLR, computation of price of house on the basis of Area of house.

MLR, computation of price of house on the basis of Area of house, Number of rooms and Parking space.

Question 24. How does polynomial regression differ from linear regression?

Answer:-> The difference of polynomial regression from linear regression.

1. The linear regression works efficiently on only those data points that are linear when plotted because in that case it becomes easier to find out the best fit line from the graph whereas in the polynomial regression since the data points are not in the linear manner when plotted, then in that case slr/mlr does'nt work efficiently and so we have to take polynomial regression into account.

2. In linear regression say in mlr, the equation changes as per the more features are added, say:

h_theta(x) = b0+b1x1+b2x2+...+bnxn

But in polynomial regression, the x remains the same as it was in the equation, just the slope changes.

h_theta(x) = b0+b1x1+b2x1^2+...+bnx1^n
